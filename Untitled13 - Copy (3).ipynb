```python
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt
import os

# Define the bandpass filter function
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    if highcut >= nyquist or lowcut <= 0:
        raise ValueError("Filter critical frequencies must be greater than 0 and less than Nyquist frequency.")
    low = lowcut / nyquist
    high = highcut / nyquist
    if low >= high:
        raise ValueError("Low frequency must be less than high frequency.")
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

# Read the CSV file
input_file = r'F:\1\112345.csv'  # Replace with your input CSV file path
df = pd.read_csv(input_file, low_memory=False) 

# Define the sampling frequency and the bandpass ranges
fs = 1700  # Replace with your actual sampling frequency
bandpass_ranges = [(20, 100), (100, 200), (200, 300), (300, 400), (400, 500)]

# Create an output directory if it doesn't exist
output_dir = r'F:\1\12' # Replace with your output directory path
os.makedirs(output_dir, exist_ok=True)

# Apply the bandpass filters to each row and save to separate CSV files
for idx, row in df.iterrows():
    first_row = row.values
    if not isinstance(first_row, np.ndarray):
        raise ValueError("Row data should be a numpy array.")

    filtered_data = []
    for lowcut, highcut in bandpass_ranges:
        try:
            filtered_row = bandpass_filter(first_row, lowcut, highcut, fs)
            filtered_data.append(filtered_row)
        except ValueError as e:
            print(f"Error with bandpass range ({lowcut}, {highcut}): {e}")

    # Convert the filtered data to a DataFrame
    filtered_df = pd.DataFrame(filtered_data)
    
    # Save the filtered data to a separate CSV file
    output_file = os.path.join(output_dir, f'filtered_row_{idx}.csv')
    filtered_df.to_csv(output_file, index=False, header=False)

    print(f"Filtered data for row {idx} saved to {output_file}")
```

```python
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt
import os

# Define the bandpass filter function
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    if highcut >= nyquist or lowcut <= 0:
        raise ValueError("Filter critical frequencies must be greater than 0 and less than Nyquist frequency.")
    low = lowcut / nyquist
    high = highcut / nyquist
    if low >= high:
        raise ValueError("Low frequency must be less than high frequency.")
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

# Read the CSV file
input_file = r'F:\1\112345.csv'  # Replace with your input CSV file path
df = pd.read_csv(input_file, low_memory=False) 

# Define the sampling frequency and the bandpass ranges
fs = 1700  # Replace with your actual sampling frequency
bandpass_ranges = [(20, 100)]

# Create an output directory if it doesn't exist
output_dir = r'F:\1\122' # Replace with your output directory path
os.makedirs(output_dir, exist_ok=True)

# Apply the bandpass filters to each row and save to separate CSV files
for idx, row in df.iterrows():
    first_row = row.values
    if not isinstance(first_row, np.ndarray):
        raise ValueError("Row data should be a numpy array.")

    filtered_data = []
    for lowcut, highcut in bandpass_ranges:
        try:
            filtered_row = bandpass_filter(first_row, lowcut, highcut, fs)
            filtered_data.append(filtered_row)
        except ValueError as e:
            print(f"Error with bandpass range ({lowcut}, {highcut}): {e}")

    # Convert the filtered data to a DataFrame
    filtered_df = pd.DataFrame(filtered_data)
    
    # Save the filtered data to a separate CSV file
    output_file = os.path.join(output_dir, f'filtered_row_{idx}.csv')
    filtered_df.to_csv(output_file, index=False, header=False)

    print(f"Filtered data for row {idx} saved to {output_file}")
```

```python
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt

# Define the bandpass filter function
def bandPassFilter(signal, fs=2000.0, low_cutoff=1.0, high_cutoff=100.0):
    nyq = 0.5 * fs
    low_cutoff_norm = low_cutoff / nyq
    high_cutoff_norm = high_cutoff / nyq

    # Butterworth filter design
    b, a = scipy.signal.butter(4, [low_cutoff_norm, high_cutoff_norm], btype='band')
    
    # Apply the filter to the EMG signal
    y = scipy.signal.filtfilt(b, a, signal, axis=0)
    return y

# Load the data
df = pd.read_csv(r'F:\data - Copy\arife\new\Book1.csv')

# Convert data to NumPy array
emg_signal = np.array(df['values'])  # Assuming the column name is 'values'

# Apply the bandpass filter
filtered_emg_signal = bandPassFilter(emg_signal)

# Save the filtered signal to a new CSV file
df['Filtered_EMG'] = filtered_emg_signal
df.to_csv(r'F:\data - Copy\arife\new\Bookq1.csv', index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

def bandpass_filter(data, lowcut, highcut, fs, order=5):
    """
    Applies a bandpass filter to the given data.

    Args:
        data: The input data.
        lowcut: The lower cutoff frequency.
        highcut: The upper cutoff frequency.
        fs: The sampling frequency.
        order: The filter order (default is 5).

    Returns:
        The filtered data.
    """

    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = signal.butter(order, [low, high], btype='bandpass')
    y = signal.filtfilt(b, a, data)
    return y

# Load EMG data from CSV
emg_data = pd.read_csv(r'F:\data - Copy\arife\new\Book1.csv')

# Assuming the EMG data is in a column named 'emg'
emg_signal = emg_data['values'].values

# Sampling frequency (replace with actual sampling frequency)
fs = 1700  # Example sampling frequency

# Bandpass filter parameters
lowcut = 100

highcut = 200



# Apply bandpass filter
filtered_signal = bandpass_filter(emg_signal, lowcut, highcut, fs)

# Create a new DataFrame with the filtered signal
filtered_data = pd.DataFrame({r'F:\data - Copy\arife\new\Bookq1.csv': filtered_signal})

# Save filtered data to a new CSV file
filtered_data.to_csv(r'F:\data - Copy\arife\new\Bookq1.csv', index=False)

# Optional: Plot the original and filtered signals for visualization
plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.plot(emg_signal, label='Original EMG')
plt.legend(loc='best')
plt.subplot(212)
plt.plot(filtered_signal, label='Filtered EMG')
plt.legend(loc='best')
plt.show()
```

```python
import scipy
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
def butter_bandpass(lowcut, highcut, fs, order=5, label=None):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    sos = signal.butter(order, [low, high], btype='band', output='sos')
    w, h = signal.sosfreqz(sos,worN=20000)
    plt.semilogx((fs * 0.5 / np.pi) * w, abs(h), label=label)
    return sos

center_freqs = np.array([0.5* 2 ** (n * 1 / 3.) for n in range(0, 29)])
center_freqs.sort()
lower_freqs = 2. ** (-1 / 6.) * center_freqs
for lf in lower_freqs:
    butter_bandpass(lf, lf*2**(1/3.), fs=2000, order=5)

plt.show()
```
